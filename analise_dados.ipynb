# Valores Discrepantes (Outliers)

```python
lista_idades = [26, 30, 32, 22, 26, 35, 400, 20, 43, 31, 23] #400 é o valor discrepante
```

```python
media = np.mean(lista_idades)
media
```

```python
#Identificando outliers com desvio padrão

desvio = np.std(lista_idades)
desvio
```

```python
#Confirmando os outliers com este seguinte cálculo abaixo
#Temos nosso limite superior, acima disso é tudo outlier

media + 3 * desvio
```

```python
#Temos nosso limite inferior, abaixo disso é tudo outlier

media - 3 * desvio
```

```python
#Entendendo se os limites calculados fazem sentido
#Usando a mediana e aprendendo sobre quartis

#Quartis - dividindo o dado em 4 partes usando 3 valores
#A mediana é o quartil do meio

#Primeiro quartil é a metade de baixo da mediana
#Terceiro quartil é a metade de cima da mediana
```

```python
#Visualizando isso com o matplotlib

import matplotlib.pyplot as plt
```

```python
#Visualizando os quartis

#Linha laranja = mediana
#Bordas = Quartil Q1 e Quartil Q3
#Linha de baixo = limite inferior
#Linha de cima = limite superior

#Tudo além é outlier

plt.boxplot(lista_idades)
```

```python
plt.boxplot(dados["SALARIO"])
```

```python
#Calculando limite inferior e limite superior para a coluna de salário com a regra dos quartis

#Função quantile(), que nos retornará os quartis

#O primeiro quartil pega o valor que separa 1/4 dos dados, então chamamos ele passando o parâmetro de 0.25
#O terceiro quartil pega o valor que separa 3/4 dos dados, então chamamos ele passando o parâmetro de 0.75
```

```python
#Primeiro Quartil

Q1 = dados["SALARIO"].quantile(0.25)
Q1
```

```python
#Terceiro Quartil

Q3 = dados["SALARIO"].quantile(0.75)
Q3
```

```python
#Interquartil - valor entre o primeiro quartil e o terceiro quartil

IQR = Q3 - Q1
IQR
```

```python
#Para pegar os limites
#Limite superior = Q3 + (1.5 * IQR)
#Limite inferior = Q1 - (1.5 * IQR)
```

```python
lim_superior = Q3 + (1.5 * IQR)
lim_superior
```

```python
lim_inferior = Q1 - (1.5 * IQR)
lim_inferior
```

```python
from scipy.stats import chi2_contingency
```

```python
def cramer_coeficiente(coluna1, coluna2):
  tabela_cruzada = np.array(pd.crosstab(coluna1, coluna2))
  chi2 = chi2_contingency(tabela_cruzada)[0]
  soma = np.sum(tabela_cruzada)
  mini = min(tabela_cruzada.shape)-1 #min(tabela_cruzada.shape) - isso vai pegar o valor mínimo do tamanho da tabela
  cramer = np.sqrt(chi2 / (soma * mini))
  return cramer
```

```python
cramer_coeficiente(dados["COR/RACA/ETNIA"], dados["NIVEL DE ENSINO"])
#Essas colunas não têm muita relação
#Porém sabemos das desigualdades na realidade, é muito provável que, de forma geral, quem está na área de TI já tem algum tipo de formação
```

```python
#Faça você mesme: Faça a correlação entre nível de ensino e gênero

def cramer_coeficiente_genero(coluna1, coluna2):
  tabela_cruzada_genero = np.array(pd.crosstab(coluna1, coluna2)) #Array porque quero em Matriz
  chi2_genero = chi2_contingency(tabela_cruzada_genero)[0]
  soma_genero = np.sum(tabela_cruzada_genero)
  mini_genero = min(tabela_cruzada_genero.shape)-1
  cramer = np.sqrt(chi2_genero / ( soma_genero * mini_genero))
  return cramer
```
